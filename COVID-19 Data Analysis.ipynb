**COVID-19 Data from John Hopkins University**

Coronavirus disease (COVID-19) is an infectious disease that took the world by storm early 2020. It is a respiratory disease that we have not been exposed to before. The speed at which it spread and how contagious it was quickly made it a global pandemic. As a result, life is very different now with many new safety procedures and protocols in place. Though we are working hard battling this pandemic, it is unlikely that we will return to our regular lives anytime soon.

In this analysis, we will analyze the number of confirmed cases and number of deaths within the United States as well as on a global scale. We will also look to see which countries are at the top of the list for both categories.

We see that the number of confirmed cases globally grow steadily over time which is unsurprising. But when we look at the shape of the graph for the United States, we see that it peaked shortly after April, declined for a few months before peaking again mid-July, followed by a few more months in decline before skyrocketing up after October. After the spike in April, most of the states implemented a stay-at-home order which likely contributed to the decline of cases over the next few months. It wasn't until mid-July when the cases started rising again. A possible narrative is that around that time, the Black Lives Matter movement picked up momentum as a result of the George Floyd incident. As people protested and gathered in large outdoor groups, the pandemic rises once again until the BLM movement dwindles. The spike at the end of the data could be a result of the holiday season approaching. As people are heading home to see family and friends, the number of COVID cases could rise as a result in the increase of social interactions.

While the cases have risen globally mostly linearly, fortunately, the number of deaths do not follow the same trend. The number of deaths spikes every now and then globally but doesn't increase with the number of cases. Similarly in the United States, the number of deaths fluctuates between 500 to 1,500 throughout the months. The number of cases might be increasing, but the mortality rate is decreasing. This is a great sign.
